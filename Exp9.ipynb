```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('/content/Wholesale customers data.csv')
```

```python
data.head()
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(data)
```

```python
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
```

```python
data_scaled.head()
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")
# Compute and plot the dendrogram using Ward's method
dendrogram = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.show()
# Set the figure size for the plot
plt.figure(figsize=(10, 7))
# Set the title of the plot
plt.title("Dendrograms")
dendrogram = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
# Add a horizontal line at y=6 with red dashed linestyle
plt.axhline(y=6, color='r', linestyle='--')
# Display the plot
plt.show()
# Set the figure size for the plot
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dendrogram = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y=6, color='r', linestyle='--')
plt.show()
# Set the figure size for the plot
clustering = AgglomerativeClustering(n_clusters=4).fit(data_scaled)
plt.figure(figsize=(10, 7))
plt.scatter(data_scaled['Milk'], data_scaled['Grocery'], c=clustering.labels_, cmap='viridis')
# Set labels and title
plt.xlabel('Milk')
plt.ylabel('Grocery')
plt.title('Clusters Visualization')
# Display the plot
plt.colorbar(label='Cluster')
plt.show()
```
